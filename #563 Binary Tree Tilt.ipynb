```python
import queue

class BinaryTreeNode:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None


def buildTree():
    inputData=[int(x) for x in input().strip().split()]
    
    rootData=inputData[0]
    root= BinaryTreeNode(rootData)
    
    q=queue.Queue()
    q.put(root)
    
    i=1
    while not q.empty():
        currNode=q.get()
        leftChildData=inputData[i]
        if leftChildData!=-1:
            leftChild=BinaryTreeNode(leftChildData)
            currNode.left=leftChild
            q.put(leftChild)
        
        i+=1
        rightChildData=inputData[i]
        if rightChildData!=-1:
            rightChild=BinaryTreeNode(rightChildData)
            currNode.right=rightChild
            q.put(rightChild)
        
        i+=1
    
    return root

    
    
    
def printLevelWiseTree(root):
    q=queue.Queue()
    q.put(root)
    q.put(None)
    while (not(q.empty())):
        current=q.get()
        if current != None:
            print(current.data,end=" ")
            if current.left != None:
                q.put(current.left)
            if current.right != None:
                q.put(current.right)
        else:
            if not(q.empty()):
                print()
                q.put(None)


                

def tiltTree(root):
    if root is None:
        return 0,0
    
    leftTilt,leftSum=tiltTree(root.left)
    rightTilt,rightSum=tiltTree(root.right)
    
    tilt=abs(leftSum-rightSum)
    rootSum=tilt+root.data
    return tilt,rootSum 

def returnTilt(root):
    tilt=tiltTree(root)[0]
    return tilt
    
root=buildTree()
printLevelWiseTree(root)
print()

print(returnTilt(root))
           
```
